Data exploration of the apples data over time

```python
import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns 
sns.set_style("whitegrid")
```

### Apples 1889 +

```python
# import data
url = 'https://raw.githubusercontent.com/mckenzie-maidl/data5100-climate-agriculture/main/Cleaned%20Data/WA_apples_1889_2021.csv'
apples = pd.read_csv(url)
apples.drop(columns='Unnamed: 0', inplace=True)
```

```python
apples.head()
```

```python
apples.info()
```

#### Yields

```python
plt.figure(figsize = (15,10))
plt.plot(apples["year"], apples["yield_acre_lbs"], color = '#C7372F')
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Yield per Acre (lbs)', fontsize = 14)
plt.title('Apple Yields per Acre by Year', fontsize = 15)
plt.tick_params(labelsize = 12)
```

```python
#bar
plt.figure(figsize = (15,10))
plt.bar(apples["year"], apples["yield_acre_lbs"])
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Yield per Acre (lbs)', fontsize = 14)
plt.title('Apple Yields per Acre by Year', fontsize = 15)
plt.tick_params(labelsize = 12)
```

#### Production

```python
#scatter
plt.figure(figsize = (15,10))
plt.scatter(apples["year"], apples["total_production_lbs"], color = '#C7372F')
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Total Apple Production (lbs)', fontsize = 14)
plt.title('Total Apple Production by Year', fontsize = 15)
plt.tick_params(labelsize = 12)
```

```python
#line
plt.figure(figsize = (15,10))
plt.bar(apples["year"], apples["total_production_lbs"], color = '#C7372F', alpha = 0.6)
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Total Apple Production (lbs)', fontsize = 14)
plt.title('Total Apple Production by Year', fontsize = 15)
plt.tick_params(labelsize = 12)
```

```python
# total vs utilized production 1950 - 2021
apples_1950 = apples[apples['year'] >= 1950]

plt.figure(figsize = (15,10))
plt.scatter(apples_1950["total_production_lbs"], apples_1950["utilized_production_lbs"], c = apples_1950['year'])
plt.xlabel('Total Apple Production (lbs)', fontsize = 14)
plt.ylabel('Utilized Apple Production (lbs)', fontsize = 14)
plt.title('Utilized vs Total Apple Production', fontsize = 15)
plt.tick_params(labelsize = 12)
```

```python
# new field
apples['percent_production_utlized'] = apples['utilized_production_lbs'] / apples['total_production_lbs']

plt.figure(figsize = (15,10))
plt.scatter(apples["year"], apples["percent_production_utlized"], color = '#C7372F')
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Percent Utilized', fontsize = 14)
plt.title('Apple Production: Percent Utilized', fontsize = 15)
plt.tick_params(labelsize = 12)
```

#### Price per Pound

```python
#scatter
plt.figure(figsize = (15,10))
plt.scatter(apples["year"], apples["avg_price_pound_dollars"])
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Average Price per Pound ($)', fontsize = 14)
plt.title('Average Apple Price per Pound', fontsize = 15)
plt.tick_params(labelsize = 12)
```

```python
#bar
plt.figure(figsize = (15,10))
plt.bar(apples["year"], apples["avg_price_pound_dollars"], color = '#C7372F', alpha = 0.6)
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Average Price per Pound ($)', fontsize = 14)
plt.title('Average Apple Price per Pound', fontsize = 15)
plt.tick_params(labelsize = 12)
plt.xlim(1905, 2025)
```

#### Bearing

```python
#scatter
plt.figure(figsize = (15,10))
plt.scatter(apples["year"], apples["bearing_acre"], color = '#C7372F')
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Acres', fontsize = 14)
plt.title('Apple Bearing Acres', fontsize = 15)
plt.tick_params(labelsize = 12)
```

```python
#bar
plt.figure(figsize = (15,10))
plt.bar(apples["year"], apples["bearing_acre"])
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Acres', fontsize = 14)
plt.title('Apple Bearing Acres in Washington', fontsize = 15)
plt.tick_params(labelsize = 12)
```

### Apples 1970 +

```python
# get new apple data
url_70 = 'https://raw.githubusercontent.com/mckenzie-maidl/data5100-climate-agriculture/main/Cleaned%20Data/apples_1970.csv'
apples_1970 = pd.read_csv(url_70)
apples_1970 = apples_1970.drop(columns = ['Unnamed: 0'])
apples_1970.head()
```

#### Total Yields

```python
# plot yields over time
plt.figure(figsize = (15,10))
plt.scatter(apples[apples['year'] >= 1970].year, apples[apples['year'] >= 1970].total_production_lbs, color = '#439d8a')
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Total Apple Production (billions of lbs)', fontsize = 14)
plt.title('Total Apple Production by Year', fontsize = 15)
plt.tick_params(labelsize = 12)
scale_y = 1e9
plt.gca().set_yticklabels(['{0:g}'.format(x/scale_y) for x in plt.gca().get_yticks()]) # change to billions
```

#### Yield per Acre

```python
# plot yields per acre over time
plt.figure(figsize = (15,10))
plt.bar(apples_1970['year'], apples_1970['yield_acre_lbs'], color = '#439d8a', alpha = 0.9)
plt.title('Apple Yields per Acre by Year', fontsize = 15)
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Yield per Acre (lbs)', fontsize = 14)
plt.tick_params(labelsize = 12)
plt.xlim(1969,2022)
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in plt.gca().get_yticks()]) # add thousand sep
```

```python
# get best fit line: y = a*x + b
xs = apples_1970['year']
ys = apples_1970['yield_acre_lbs']
trend = np.polyfit(xs,ys,1)
trendpoly = np.poly1d(trend) 

# scattter plot with best fit line
plt.figure(figsize = (15,10))
plt.plot(xs,trendpoly(xs), color = '#439d8a')
plt.plot(xs,ys,'o', color = '#439d8a')
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Yield per Acre (lbs)', fontsize = 14)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in plt.gca().get_yticks()])
plt.xlim(1969,2022)
```

#### Changes to Production

```python
# trees per acre and yield per acre
fig, ax1 = plt.subplots(figsize = (15,10))
ax2 = ax1.twinx()
ax2.grid(False)

ax1.scatter(apples_1970['year'], apples_1970['yield_acre_lbs'], color = '#439d8a')
ax1.set_ylabel('Yield per Acre (lbs)', fontsize = 14, color = '#439d8a')
ax1.get_yaxis().set_major_formatter(mpl.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
ax1.tick_params(labelsize = 12)

ax1.set_xlabel('Year', fontsize = 14)
ax1.set_xlim(1969,2022)

ax2.bar(apples_1970['year'], apples_1970['trees_per_acre'], color = '#40b4e5', alpha = 0.2)
ax2.set_ylabel('Trees per Acre', fontsize = 14, color='#40b4e5')
ax2.tick_params(labelsize = 12)

plt.show()
```

```python
# yield per tree
apples_1970['yield_tree'] = apples_1970['yield_acre_lbs']/apples_1970['trees_per_acre']

sns.set_style("whitegrid")
plt.figure(figsize = (15,10))
plt.scatter(apples_1970['year'], apples_1970['yield_tree'], color = '#439d8a')
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Yield per Tree (lbs)', fontsize = 14)
plt.tick_params(labelsize = 12)
plt.xlim(1969,2022)
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in plt.gca().get_yticks()]) # add thousand sep
```

#### Percent Utilized

```python
plt.figure(figsize = (15,10))
plt.scatter(apples_1970['year'], apples_1970['percent_utilized'], color = '#439d8a')
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Percent of Production Utilized', fontsize = 14)
plt.tick_params(labelsize = 12)
plt.xlim(1969,2022)
```

```python
# all utilized
apples_1970['all_utilized'] = np.where(apples_1970['percent_utilized'] < 100.000000, 0, 1)

plt.figure(figsize = (15,3))
plt.scatter(apples_1970['year'], apples_1970['all_utilized'], color = '#439d8a')
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Total Production Utilized', fontsize = 14)
plt.tick_params(labelsize = 12)
plt.yticks([0,1])
plt.xticks(np.arange(1970, 2022, 5))
plt.xlim(1969,2022)
```
